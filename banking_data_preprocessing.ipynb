```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv(r'C:\Users\vshal\Downloads\Banking-Transaction-Risk-Financial-Reporting-System\data\raw\bank_transactions_data_.csv')
```

```python
# Show first rows
print(df.head())
```

```python
# Convert date columns
df['TransactionDate'] = pd.to_datetime(
    df['TransactionDate'],
    dayfirst=True
)

df['PreviousTransactionDate'] = pd.to_datetime(
    df['PreviousTransactionDate'],
    dayfirst=True
)
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
# Remove missing values
df = df.dropna()
```

```python
# Create transaction category
df['TransactionCategory'] = df['TransactionAmount'].apply(
    lambda x: 'High' if x > 5000 else 'Normal'
)
```

```python
# Create risk score
df['RiskScore'] = (
    df['LoginAttempts'] * 10
    + (df['TransactionAmount'] / 100)
)
```

```python
# Save cleaned data
df.to_csv(
    r'C:\Users\vshal\Downloads\Banking-Transaction-Risk-Financial-Reporting-System\data\processed\cleaned_transactions.csv',
    index=False
)

print("ETL Process Completed Successfully")
```
